### importando os dados dos exemplares


```python
dados_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
```

### Concatenando os dataframes de emprestimos

### Unindo os DataFrames
unir os dataFrames de emprestimo e exemplares


```python
dados_biblioteca_completos = emprestimos_biblioteca.merge(dados_exemplares)
dados_biblioteca_completos
```

```python
dados_biblioteca_completos.columns
```

Verificar nulos

```python

dados_biblioteca_completos.isnull().sum()
```

```python
# verificar os valores unicos na coluna matricula_ou_siape
#dados.groupby('GENERO', dropna=False)['ID'].nunique()
dados_biblioteca_completos.groupby('matricula_ou_siape', dropna=False)['id_emprestimo'].nunique()
```

```python
# filtrar os dados para verificar se matricula_ou_siape que não esta preenchida agrupando com tipo_vinculo_usuario

dados_biblioteca_completos[dados_biblioteca_completos['matricula_ou_siape'].isnull()].groupby('tipo_vinculo_usuario', dropna=False)['id_emprestimo'].nunique()
```
